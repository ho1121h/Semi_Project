```python
import json
import re
import urllib.request as req

def to_html(i):
    url = f'https://store.steampowered.com/search/results/?query&start={i}&count=100&dynamic_data=&force_infinite=1&category1=998&snr=1_7_7_230_7&infinite=13'
    res = req.urlopen(url)
    json_obj = json.load(res)
    json_obj
    html=json_obj['results_html']
    return html
    
def appid_title(n):
    appid_title=[]
    for i in (100,n+1,100):
        herf =re.findall('store.steampowered.com/app.+',to_html(i))
        for j in herf:
            j=j[27:]
            j=j[:j.find('?')-1]
            appid_title.append(j.split('/'))
    return appid_title

appid_title(00)
```
