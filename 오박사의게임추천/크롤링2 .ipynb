```python
import time
import requests
import csv
from selenium import webdriver
from bs4 import BeautifulSoup
import os
import sys
from selenium.webdriver.common.keys import Keys
import chromedriver_autoinstaller
import warnings
```

```python
warnings.filterwarnings('ignore')
#셀레니움 사용
options = webdriver.ChromeOptions()
options.add_argument('window-size=1280,800')

b = webdriver.Chrome('chromedriver', options=options)
b.implicitly_wait(5)

b.get(url='https://store.steampowered.com/search/?filter=topsellers')
b.implicitly_wait(10)
```

```python
평가_list = []
title_list = []
id_list = []
for i in range(1,2) :#모든 내용 로드를 위한 동작
    info_n = b.execute_script("return document.body.scrollHeight")#로드된 내용의 최하단 크기확인
    b.execute_script("window.scrollTo(0, document.body.scrollHeight)")#로드된 내용의 최하단으로 이동
    # 확장
    b.implicitly_wait(5)
    next_n = b.execute_script("return document.body.scrollHeight")
    if info_n == next_n:
        break
    
    
    article_raw = b.find_elements_by_class_name('search_result_row')
    # 제목 크롤링 시작
    for article in article_raw:
        
        title = article.find_element_by_class_name('title').text
        title_list.append(title)
        
        try:
            평가 =article.find_element_by_class_name("search_review_summary").get_attribute("data-tooltip-html")
            평가_list.append(평가)
        except:
            평가_list.append("")
        
        try:
            gameid = article.get_attribute("data-ds-appid")
            id_list.append(gameid) 
        except:
            id_list.append("")
    print('평가 갯수: ', len(평가_list))
    print('title 갯수: ', len(title_list))
    print('id 갯수: ', len(id_list))
```

```python
import pandas as pd
import numpy as np
# 수집된 url_list, title_list로 판다스 데이터프레임 만들기
df = pd.DataFrame({'매출 순위':title_list,'평가':평가_list,'게임번호':id_list})
df#1.스플릿 으로 나눠보자
```

```python
df.to_csv("steam.csv", encoding='utf-8-sig')
```
