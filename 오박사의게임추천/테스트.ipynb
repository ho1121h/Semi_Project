```python
from sklearn.feature_extraction.text import CountVectorizer
import pandas as pd  # 데이터 분석을 위한 라이브러리
import numpy as np   # 넘파이 : 숫자 계산

# 워닝 끄기
import warnings
warnings.filterwarnings('ignore')
```

```python
games = pd.read_csv('리뷰.csv')
#games.gameid.values
```

```python
from sentence_transformers import SentenceTransformer
corpus = games['review'].values.tolist() # 리뷰 임베딩

model = SentenceTransformer('jhgan/ko-sroberta-multitask')

embeddings = model.encode(corpus)# 리뷰 데이터 전부 인코딩 (센텐스 버트 모델)

embeddings[:5]
#임베딩 = 정규화(컴퓨터가 알아먹게끔)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
games_sim = cosine_similarity(embeddings, embeddings)

print(games_sim.shape)
print(games_sim[:10])
```

```python
games_sim_sorted_ind = games_sim.argsort()[:, ::-1] # ::-1 : 역순으로 정렬
print(games_sim_sorted_ind[:5])
```

```python
def find_sim_games_ver1(df, sorted_ind, title_name, top_n=5):
    
    # 인자로 입력된 movies_df DataFrame에서 'title' 컬럼이 입력된 title_name 값인 DataFrame추출
    title_games = df[df['gameid'] == title_name]
    
    # title_named을 가진 DataFrame의 index 객체를 ndarray로 반환하고 
    # sorted_ind 인자로 입력된 genre_sim_sorted_ind 객체에서 유사도 순으로 top_n 개의 index 추출
    title_index = title_games.index.values
    similar_indexes = sorted_ind[title_index, :(top_n)]
    
    # 추출된 top_n index들 출력. top_n index는 2차원 데이터 임.
    # dataframe에서 index로 사용하기 위해서 1차원 array로 변경
    print(similar_indexes)    
    # 2차원 데이터를 1차원으로 변환
    similar_indexes = similar_indexes.reshape(-1)
    
    return df.iloc[similar_indexes]
```

```python
similar_games = find_sim_games_ver1(games, games_sim_sorted_ind, 'ELDEN_RING', 5)

similar_games[['gameid']].drop_duplicates()
```
